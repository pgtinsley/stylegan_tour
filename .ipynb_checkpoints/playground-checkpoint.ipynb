```python
import cv2

import os
import pickle
import PIL.Image
import numpy as np
import dnnlib
import dnnlib.tflib as tflib
import config
from encoder.generator_model import Generator

import face_recognition

import matplotlib.pyplot as plt
%matplotlib inline

import random
```

```python
def setup():
    tflib.init_tf()
    # Load pre-trained network.
    url = 'https://drive.google.com/uc?id=1MEGjdvVpUsu1jB4zrXZN7Y4kBBOzizDQ' # karras2019stylegan-ffhq-1024x1024.pkl
    with dnnlib.util.open_url(url, cache_dir=config.cache_dir) as f:
        _G, _D, Gs = pickle.load(f)
    generator = Generator(Gs, batch_size=1, randomize_noise=False) # -- RUNNING >1 TIMES THROWS ERROR
    fmt = dict(func=tflib.convert_images_to_uint8, nchw_to_nhwc=True)
    return [_G, _D, Gs, generator, fmt]
```

```python
# Only run once.
[_G, _D, Gs, generator, fmt] = setup()
```

```python
def generate_image(latent_vector):
    latent_vector = latent_vector.reshape((1, 18, 512))
    generator.set_dlatents(latent_vector)
    img_array = generator.generate_images()[0]
    img = PIL.Image.fromarray(img_array, 'RGB')
    return img.resize((256, 256))
```

```python
def change_identities(vec1, vec2, fname='change_identities.avi'):
    vec1_copy = np.copy(vec1)
    images = []
    # Sub out components one-by-one (coarse->medium->fine)
    for i in range(0, 18):
        vec1_copy[i] = vec2[i] # do the actual substitution 
        image = np.array(generate_image(vec1_copy)) # generate image
        images.append(image)
    s = images[0].shape
    videowriter =  cv2.VideoWriter(fname, cv2.VideoWriter_fourcc(*'mp4v'), 6, (s[1], s[0]))
    for i in range(0, 18):
        videowriter.write(images[i][...,::-1])
    videowriter.release()
```
